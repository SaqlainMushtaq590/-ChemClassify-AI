```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
x = df['message_1'].astype(str).values
y = df['topic;'].values
print(f"Total Samplsequence of numbere Loaded:{len(y)}")
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.preprocessing import LabelEncoder

# Data processing
max_words = 5000
tokenizer = Tokenizer(num_words=max_words, oov_token="<oov>")
tokenizer.fit_on_texts(x)
sequences = tokenizer.texts_to_sequences(x)

# Padding
max_length = 100 
x_padded = pad_sequences(sequences, maxlen=max_length, padding="post")

# Encoding
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Output
num_classes = len(label_encoder.classes_)
print(f"Unique Classes Identified: {num_classes}")
```

```python
print(f"Unique Classes Identified: {y_encoded}")
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_padded, y_encoded, test_size= 0.3, random_state = 42) 
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(input_dim = max_words, output_dim =64),
    tf.keras.layers.Conv1D(filters = 128, kernel_size= 5, activation = "relu"),
    tf.keras.layers.GlobalMaxPooling1D(),
    tf.keras.layers.Dense(64, activation = "relu"),
    tf.keras.layers.Dense(num_classes, activation= "softmax")
])

model.compile(optimizer = "adam",
              loss = "sparse_categorical_crossentropy",
              metrics = ["accuracy"])
model.build(input_shape = (None, max_length))
model.summary()
```

```python
history = model.fit(x_train, y_train,
                    epochs =10,
                    validation_data = (x_test, y_test),
                    batch_size = 64)
```

```python
def pred_topic(new_question):
    seq = tokenizer.texts_to_sequences([new_question])
    padded = pad_sequences(seq, maxlen = max_length)
    prediction = model.predict(padded, verbose = 0)
    class_index = np.argmax(prediction)
    return label_encoder.classes_[class_index]
```

```python
test_q = "Explain the process of Lucas Test."
print(f"Prediction: {pred_topic(test_q)}")
```

```python
test_q = "egg gets rotten"
print(f"Prediction: {pred_topic(test_q)}")
```
